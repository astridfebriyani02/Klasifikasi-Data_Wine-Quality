Data tersebut merupakan data yang menggambarkan karakteristik kimia dan kualitas dari sampel wine yang telah diuji. Dataset ini sering digunakan untuk melakukan analisis dan prediksi terhadap tingkat kualitas wine berdasarkan kandungan kimia yang dimilikinya. Berikut adalah variabel-variabel yang terdapat dalam dataset:

*   fixed acidity : kadar asam non-volatil (seperti asam tartarat) yang tetap setelah proses fermentasi.
*   volatile acidity : kadar asam volatil (seperti asam asetat) yang dapat menguap dan jika terlalu tinggi dapat menyebabkan aroma tidak sedap.
*   citric acid : kandungan asam sitrat yang dapat memberikan rasa segar pada wine.
*   residual sugar : jumlah gula yang tersisa setelah fermentasi selesai, berpengaruh pada tingkat kemanisan wine.
*   chlorides : kandungan garam (natrium klorida) dalam wine, yang mempengaruhi rasa.
*   free sulfur dioxide : jumlah sulfur dioksida bebas yang bertindak sebagai pengawet dan mencegah oksidasi.
*   total sulfur dioxide : total kadar sulfur dioksida (bebas dan terikat) dalam wine.
*   density : kerapatan cairan wine, yang berkorelasi dengan kandungan gula dan alkohol.
*   pH : tingkat keasaman wine; semakin rendah pH, semakin asam wine tersebut.
*   sulphates : senyawa yang berperan sebagai agen antimikroba dan dapat meningkatkan kestabilan wine.
*   alcohol : persentase kandungan alkohol dalam wine.
*   quality : skor kualitas wine yang diberikan oleh panel ahli (biasanya dalam skala 0–10).
*   Id : nomor identifikasi unik untuk setiap sampel wine.

Data ini mencakup berbagai variabel kimia dan sensorik yang berpengaruh terhadap mutu wine. Melalui analisis data ini, kita dapat mengidentifikasi hubungan antara karakteristik kimia dengan tingkat kualitas wine, serta mengembangkan model prediksi untuk menentukan kualitas wine berdasarkan komposisi kimianya. # **8. Modeling dan Evaluasi**

Pada analisis ini, Random Forest diterapkan untuk memodelkan kualitas wine berdasarkan karakteristik kimia dan fisika.

```python
# === Metrics & Model Selection ===
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score
from sklearn.model_selection import train_test_split

# === Model ===
from sklearn.ensemble import RandomForestClassifier
```

**Data Splitting**

```python
X=wine.drop(columns=['quality', 'Id'])
y=wine['quality']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    random_state=0,
    stratify=y
)

print("Kelas di y_train:", np.unique(y_train))
print("Kelas di y_test:", np.unique(y_test))
```

```python
# Deskripsi singkat data training
print("Info Data Training:")
print(X_train.info())

# Deskripsi singkat data testing
print("Info Data Testing:")
print(X_test.info())
```

Dari deskripsi data training dan testing di atas, dapat dilihat bahwa:

Data training terdiri dari 685 entri, sedangkan data testing terdiri dari 172 entri.
Kedua dataset memiliki 11 kolom yang bertipe float64.
Tidak ada nilai yang hilang (null) pada kedua dataset

**Modelling**

```python
# Import Library Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(n_estimators=400, random_state=42)
random_forest.fit(X_train, y_train)
Y_predrf = random_forest.predict(X_test)
accuracy_rf=round(accuracy_score(y_test,Y_predrf)* 100, 2)
acc_random_forest = round(random_forest.score(X_train, y_train) * 100, 2)

cm = confusion_matrix(y_test, Y_predrf)
accuracy = accuracy_score(y_test,Y_predrf)
precision = precision_score(y_test, Y_predrf, average='weighted', zero_division=0)
recall = recall_score(y_test, Y_predrf, average='weighted', zero_division=0)
f1 = f1_score(y_test,Y_predrf, average='weighted', zero_division=0)

print('accuracy_random_Forest : %.3f' % accuracy)
print('precision_random_Forest : %.3f' % precision)
print('recall_random_Forest : %.3f' % recall)
print('f1-score_random_Forest : %.3f' % f1)

```

Dari hasil klasifikasi menggunakan random forest diperoleh accuracy sebesar 66.9%, precision 64.9%, recall 66.9%, dan sebesar F1 score 65.4%,

```python
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
import numpy as np

# Ambil kelas unik dari y_test
classes = np.unique(y_test)

# Ubah y_test ke bentuk biner untuk multiclass ROC
y_test_bin = label_binarize(y_test, classes=classes)
n_classes = y_test_bin.shape[1]

# Memastikan y_score dihitung
y_score = random_forest.predict_proba(X_test)

# Hitung ROC curve dan ROC area untuk setiap kelas
fpr, tpr, roc_auc = dict(), dict(), dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Hitung macro-average ROC
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
mean_tpr = np.zeros_like(all_fpr)
for i in range(n_classes):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
mean_tpr /= n_classes
roc_auc["weighted"] = auc(all_fpr, mean_tpr)
```

```python
from itertools import cycle

# --- Plot ROC Curve ---
plt.figure(figsize=(8, 6))
colors = cycle(['aqua', 'darkorange', 'cornflowerblue', 'green', 'red', 'purple', 'pink'])

for i, color in zip(range(n_classes), colors):
    plt.plot(
        fpr[i],
        tpr[i],
        color=color,
        lw=2,
        label=f'ROC curve class {classes[i]} (area = {roc_auc[i]:0.2f})'
    )

plt.plot(all_fpr, mean_tpr, color='navy', linestyle='--', label=f'weighted-average ROC (area = {roc_auc["weighted"]:0.2f})', lw=3)

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multiclass ROC Curve - Random Forest')
plt.legend(loc='lower right')
plt.grid(alpha=0.3)
plt.show()
```

Untuk evaluasi lebih mendalam menggunakan ROC-AUC, model memiliki skor ROC-AUC keseluruhan sebesar 0,86, yang mengindikasikan kemampuan model dalam membedakan antar kelas secara umum cukup baik. Jika dilihat per kelas, model menunjukkan performa yang bervariasi: kelas 3 memiliki ROC-AUC tertinggi 0,95, menunjukkan model sangat baik dalam membedakan kelas ini dari kelas lainnya. Kelas 4 dan 5 memiliki skor masing-masing 0,84 dan 0,85, kelas 6 lebih rendah sebesar 0,75, sedangkan kelas 7 dan 8 masing-masing 0,90. Variasi ini mengindikasikan bahwa model lebih kuat dalam mengenali beberapa kelas tertentu, sementara kelas 6 relatif lebih sulit untuk diprediksi dibanding kelas lain.

# **9. Prediksi**

Pada tahap prediksi, data testing diproses dengan menghapus kolom Id karena bukan termasuk fitur model. Model Random Forest yang sudah dilatih sebelumnya di-load kembali, kemudian digunakan untuk memprediksi kualitas wine pada data testing. Hasil prediksi disimpan dalam DataFrame predicted_quality dan digabungkan dengan kolom Id untuk mengaitkan setiap prediksi dengan record aslinya. Hasil akhir kemudian disimpan ke dalam file CSV hasilprediksi_041.csv, dan beberapa baris pertama ditampilkan untuk memastikan prediksi telah dilakukan dengan benar.

```python
import joblib

# Simpan model Random Forest
joblib.dump(random_forest, 'model_random_forest.pkl')

# --- 1. Membaca data testing ---
dataset2 = "/content/data_testing.csv"
testing = pd.read_csv(dataset2)

print(testing.info())
display(testing.head())

# --- 2. Load model ---
loaded_model = joblib.load('model_random_forest.pkl')

# --- 3. Prediksi tanpa kolom target (quality) ---
testing_for_prediction = testing.drop(columns=['Id'], errors='ignore')
y_pred = loaded_model.predict(testing_for_prediction)

# --- Buat DataFrame dengan hasil prediksi langsung ---
hasil_prediksi = pd.DataFrame(y_pred, columns=['predicted_quality'])

# --- Gabungkan hasil prediksi dengan kolom 'Id' ---
hasil_final = pd.concat([testing[['Id']], hasil_prediksi], axis=1)

# --- Simpan ke CSV ---
hasil_final.to_csv('hasilprediksi_041.csv', index=False)
hasil_final.head()
```
